```python
import spacy 
from spacy import displacy
nlp = spacy.load('en_core_web_md')
doc = nlp("I live in New Yourk, Bruno likes studying his favorite subjects like natural language processing and machine learning ")
```

```python
# Visual lizing the data
# Test cell
from IPython.display import display
from spacy import displacy

def visualize(doc, style):
    return display(displacy.render(doc, style=str(style)))
```

```python
for style in ['ent','dep']:
    visualize(doc, style)


```
